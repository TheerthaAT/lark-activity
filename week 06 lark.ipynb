# 2. Data Visualizations

# 1. Graph of top 10 countries based on population 

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='Population in thousands (2017)',data=data.sort_values('Population in thousands (2017)',ascending=False).head(10),color='black')
plt.title('Top 10 countries based on Population in thousands (2017)',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Population in thousands (2017)',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

# 2. Graph of top 10 countries based on surface area

```python
plt.figure(figsize=(10,8))
sns.barplot(x="country" , y="Surface area (km2)" ,data=data.sort_values("Surface area (km2)" ,ascending=False).head(10) ,color="blue")
plt.title("Top 10 countries based on Surface area(km2)" , fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Surface area(km2) ',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

# 3.Graph of least 10 countries based on International trade: Imports (million US$)

```python
data1=data.nsmallest(10,["International trade: Imports (million US$)"])
data1                     
```

```python
plt.figure(figsize=(10,8))
x=data1["country"]
y=data1["International trade: Imports (million US$)"]
plt.plot(x,y)
plt.title("Least 10 countries based on International trade: Imports (million US$)" , fontsize=15)
plt.xlabel("country" , fontsize=13)
plt.ylabel("International trade: Imports (million US$)" , fontsize=13)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

# 4.Graph of least 5 countries based on  GDP per capita (current US$)

```python
plt.figure(figsize=(10,8))
sns.barplot(x="country" , y="GDP per capita (current US$)" ,data=data.sort_values("GDP per capita (current US$)",ascending=True).head(5) ,color="red")
plt.title("Least 5 countries based on GDP per capita (current US$)" , fontsize=15)
plt.xlabel("country" , fontsize=13)
plt.ylabel("GDP per capita (current US$)", fontsize=13)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

# 3. Countries with respect to regions

```python
data.groupby("Region")[["country"]].count()
```

# 4.On average which region has the highest


# a.GDP per capita(current US$)

```python
top_GDP=data.groupby("Region")[["GDP per capita (current US$)"]].mean()
top_GDP.sort_values("GDP per capita (current US$)", ascending=False)
```

```python
top_GDP.nlargest(1,"GDP per capita (current US$)")
```

# 4.b International trade:imports(million US$)

```python
top_ITI=data.groupby("Region")[["International trade: Imports (million US$)"]].mean()
top_ITI.sort_values("International trade: Imports (million US$)" , ascending=False)
```

```python
top_ITI.nlargest(1,"International trade: Imports (million US$)")
```

# 5. Consistent region with respect to GDP per capita(current US$)

```python
GDP_data=data.groupby("Region")[["GDP per capita (current US$)"]].std()
GDP_data.sort_values("GDP per capita (current US$)" , ascending=False)
```

```python
GDP_data.nsmallest(1,"GDP per capita (current US$)")
```
