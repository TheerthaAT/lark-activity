```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(data['CO AQI'], data['NO2 AQI'])
ax.set_xlabel('CO AQI')
ax.set_ylabel('NO2 AQI')
plt.show()
```

```python
data.columns
```

```python
sns.FacetGrid(data, hue="State", size=5) \
   .map(plt.scatter, "SO2 Mean", "NO2 Mean") \
   .add_legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='City',y='SO2 AQI',data=data.sort_values('SO2 AQI',ascending=False).head(15),color='black')
plt.title('Top 15 cities based on SO2 AQI',fontsize=15)
plt.xlabel('City',fontsize=15)
plt.ylabel('SO2 AQI',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
ax = sns.barplot(x="O3 Mean",y="State", data=data, ci=100)
fig = plt.gcf()
fig.set_size_inches(18, 13)
plt.show()
```

```python
plt.figure(figsize=(10,8))
x=data["Year"]
y=data["CO AQI"]
plt.plot(x,y)
plt.title("Year vs CO AQI" , fontsize=15)
plt.xlabel("Year" , fontsize=13)
plt.ylabel("CO AQI" , fontsize=13)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,8))
x=data["Year"]
y=data["SO2 AQI"]
plt.plot(x,y)
plt.title(" SO2 AQI over Time" , fontsize=15)
plt.xlabel("Year" , fontsize=13)
plt.ylabel("SO2 AQI" , fontsize=13)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
x=data["State"]
y=data["NO2 AQI"]
plt.plot(x,y)
plt.title("State vs NO2 AQI" , fontsize=15)
plt.xlabel("State" , fontsize=13)
plt.ylabel("NO2 AQI" , fontsize=13)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
